```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report

# Load the dataset
# Make sure to replace 'path_to_your_data' with the actual path to your CSV file.
df = pd.read_csv('/content/OnlineNewsPopularity.csv')

# Selecting features for the model (from our earlier discussion)
features = [
    ' n_tokens_title',
    ' n_tokens_content',
    ' n_unique_tokens',
    ' n_non_stop_words',
    ' num_hrefs',
    ' num_imgs',
    ' num_videos',
    ' average_token_length',
    ' num_keywords',
    ' global_sentiment_polarity',
    ' global_rate_positive_words'
]
# Assuming ' shares' is the target column
# Create a binary classification target: 1 if ' shares' > 1400, else 0
df['popular'] = (df[' shares'] > 1400).astype(int)

# Create the features matrix (X) and the target vector (y)
X = df[features]
y = df['popular']

# Split the dataset into a training set and a test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize the Decision Tree Classifier
decision_tree = DecisionTreeClassifier(random_state=42)

# Fit the classifier to the training data
decision_tree.fit(X_train_scaled, y_train)

# Predict on the test set
y_pred = decision_tree.predict(X_test_scaled)
y_pred_proba = decision_tree.predict_proba(X_test_scaled)[:, 1]

# Calculate the accuracy and the ROC AUC score
accuracy = accuracy_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_proba)

# Generate a classification report
report = classification_report(y_test, y_pred)

print(f'Accuracy of Decision Tree: {accuracy:.2f}')
print(f'ROC AUC Score of Decision Tree: {roc_auc:.2f}')
print(report)
```

```python
import matplotlib.pyplot as plt

# Create a simple decision tree visualization
def draw_tree():
    # Setup the figure and axis
    fig, ax = plt.subplots(figsize=(12, 6))  # Larger figure size
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    plt.axis('off')

    # Define the boxes for decision nodes and leaves
    decision_box = dict(boxstyle="round4,pad=0.6", fc="cornflowerblue", ec="black", lw=2)
    leaf_box = dict(boxstyle="round,pad=0.4", fc="lightgreen", ec="black", lw=2)

    # Node positions
    nodes = {
        "Title Length <= 9.5": (0.5, 0.8),
        "Has Images <= 5": (0.7, 0.6),
        "Has Videos <= 2": (0.7, 0.4)  # Adjusted position for clarity
    }
    # Leaf positions and labels
    leaves = {
        "Popular": [(0.3, 0.6), (0.8, 0.6)],
        "Not Popular": [(0.5, 0.3)]  # Adjusted position for clarity
    }

    # Draw nodes
    for text, (x, y) in nodes.items():
        ax.text(x, y, text, ha="center", va="center", bbox=decision_box, fontsize=14)

    # Draw leaves
    for outcome, positions in leaves.items():
        for (x, y) in positions:
            ax.text(x, y, outcome, ha="center", va="center", bbox=leaf_box, fontsize=14)

    # Draw arrows and labels
    ax.annotate('', xy=(0.3, 0.6), xytext=(0.5, 0.8),
                arrowprops=dict(arrowstyle="->", color="black", lw=2))
    ax.annotate('', xy=(0.7, 0.6), xytext=(0.5, 0.8),
                arrowprops=dict(arrowstyle="->", color="black", lw=2))
    ax.annotate('', xy=(0.5, 0.3), xytext=(0.7, 0.6),
                arrowprops=dict(arrowstyle="->", color="black", lw=2))
    ax.annotate('', xy=(0.8, 0.6), xytext=(0.7, 0.6),
                arrowprops=dict(arrowstyle="->", color="black", lw=2))
    ax.annotate('', xy=(0.8, 0.6), xytext=(0.7, 0.4),
                arrowprops=dict(arrowstyle="->", color="black", lw=2))

    # Labels for the decision paths
    ax.text((0.5 + 0.3) / 2, (0.8 + 0.6) / 2 - 0.05, "No", ha="center", fontsize=12)
    ax.text((0.5 + 0.7) / 2, (0.8 + 0.6) / 2 - 0.05, "Yes", ha="center", fontsize=12)
    ax.text((0.7 + 0.5) / 2, (0.6 + 0.3) / 2, "No", ha="center", fontsize=12)
    ax.text((0.7 + 0.8) / 2, (0.6 + 0.6) / 2 - 0.05, "Yes", ha="center", fontsize=12)
    ax.text((0.7 + 0.8) / 2, (0.4 + 0.6) / 2 - 0.05, "Yes", ha="center", fontsize=12)

    plt.show()

# Draw the decision tree
draw_tree()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('/content/OnlineNewsPopularity.csv')  # Adjust the file path accordingly

# Define the features and the target variable
X = df[[' global_sentiment_polarity']]  # Feature matrix with only content sentiment
y = (df[' shares'] > 1400).astype(int)  # Binary target variable

# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Decision Tree Classifier
decision_tree = DecisionTreeClassifier(random_state=42)

# Fit the classifier to the training data
decision_tree.fit(X_train, y_train)

# Make predictions on the test set
y_pred = decision_tree.predict(X_test)
y_pred_proba = decision_tree.predict_proba(X_test)[:, 1]

# Calculate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show()

# Calculate and print the classification report
print(classification_report(y_test, y_pred))

# Calculate ROC AUC score and plot the ROC curve
roc_auc = roc_auc_score(y_test, y_pred_proba)
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

plt.figure()
plt.plot(fpr, tpr, label=f'Decision Tree (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], linestyle='--', label='No Skill')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import GBTClassifier
from pyspark.ml.evaluation import BinaryClassificationEvaluator
from pyspark.sql.functions import col

# Initialize Spark Session
spark = SparkSession.builder.appName("ArticlePopularityByDay").getOrCreate()

# Load the data
data_path = '/content/OnlineNewsPopularity.csv'
df = spark.read.csv(data_path, header=True, inferSchema=True)

# Remove whitespaces from column names
df = df.select([col(c).alias(c.strip()) for c in df.columns])

# Select day of the week columns
day_of_week_cols = ['weekday_is_monday', 'weekday_is_tuesday', 'weekday_is_wednesday',
                    'weekday_is_thursday', 'weekday_is_friday', 'weekday_is_saturday',
                    'weekday_is_sunday', 'is_weekend']

# Cast to float required by VectorAssembler
for col_name in day_of_week_cols:
    df = df.withColumn(col_name, col(col_name).cast('float'))

# Define the label column
shares_threshold = df.approxQuantile('shares', [0.5], 0.05)[0]
df = df.withColumn('label', (col('shares') > shares_threshold).cast('integer'))

# Assemble features into a vector
assembler = VectorAssembler(inputCols=day_of_week_cols, outputCol="features")
df = assembler.transform(df)

# Split the dataset
(train_data, test_data) = df.randomSplit([0.7, 0.3])

# Train the GBT model
gbt = GBTClassifier(labelCol="label", featuresCol="features", maxIter=10)
model = gbt.fit(train_data)

# Make predictions
predictions = model.transform(test_data)

# Evaluate the model
evaluator = BinaryClassificationEvaluator()
auc = evaluator.evaluate(predictions, {evaluator.metricName: "areaUnderROC"})

# Feature Importance
importances = model.featureImportances.toArray()
importance_dict = dict(zip(day_of_week_cols, importances))

print(f"Area Under ROC: {auc}")
print("Feature Importances:")
for day, importance in sorted(importance_dict.items(), key=lambda x: x[1], reverse=True):
    print(f"{day}: {importance}")

# Stop the Spark session
spark.stop()
```
